## Calling for the data and prices of crypto using the website's API. The automated process takes up to 333 times as permitted by the API.


```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json

url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {
  'start':'1',
  'limit':'15',
  'convert':'USD'
}
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': '507b2de8-7dc3-41cd-8052-ae75db13f579',
}

session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data = json.loads(response.text)
  print(data)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```

```python
def api_runner():
    global df 
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
      'start':'1',
      'limit':'15',
      'convert':'USD'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': '507b2de8-7dc3-41cd-8052-ae75db13f579',
    }

    session = Session()
    session.headers.update(headers)

    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
      print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)


    #to keep it in a dataframe
    df2 = pd.json_normalize(data['data'])
    df2['Timestamp'] = pd.to_datetime('now')
    df = df.append(df2)



        # Use this if you want to create a csv and append data to it
    df = pd.json_normalize(data['data'])
    df['timestamp'] = pd.to_datetime('now')
    df.head()

    if not os.path.isfile(r'D:\workspace\portfolio projects\py project2\API.csv'):
        df.to_csv(r'D:\workspace\portfolio projects\py project2\API.csv', header='column_names')
    else:
        df.to_csv(r'D:\workspace\portfolio projects\py project2\API.csv', mode='a', header=False)
```

```python
import os 
from time import time
from time import sleep
#the API documentation said it grants 333 runs so the range can be up to 333 but we keepin it 100
for i in range(100):
    api_runner()
    print('API Runner completed')
    sleep(60) #sleep for 1 minute
exit()
```
