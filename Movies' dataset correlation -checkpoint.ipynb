# Movie Industry Dataset Analysis
#### Link to [dataset](https://www.kaggle.com/datasets/danielgrijalvas/movies)

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# to read in the data
df = pd.read_csv('movies.csv')
```

```python
df.head(3)
```

```python
#Explore the shape of the dataset 

shape  = df.shape
print(shape) 

df.info()
```

Dataset initially had 7668 rows and 15 columns

```python
#checks for null values
df.isna().sum()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for columns

print(df.dtypes)
```

```python
df.drop_duplicates()
shape = df.shape
print(shape)
```

Dataset has no duplicate entries

```python
#finding the highest movies in earnings 
df = df.sort_values(by=['gross'],ascending = False, inplace=False) 
df.head()
```

```python
#plotting budget vs gross using seaborn to find correlation 
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
df.corr(method='pearson')
```

There is a high correlation between budget and gross

```python
#visualizing the correlation 
Correlation_matrix = df.corr(method='pearson')
```

```python
sns.heatmap(Correlation_matrix, annot=True)
plt.title('correlation matrix')
plt.show()
```

```python
CompanyGrossSum = df.groupby(['company', 'year'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company','year'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
CompanyGrossSum = df.groupby(['company'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
